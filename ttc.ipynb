# Thacker's Tiny Computer 3

Thacker's Tiny Computer (_ttc_) is a very simple RISC like CPU design invented as a teaching aid.  Here we demonstrate an implementation of _ttc_ in OCaml ported from code provided by the [University of Cambridge](http://www.cl.cam.ac.uk/teaching/1314/ECAD+Arch/labs/background/ttc.html).

Load the libraries and install the notebook printers.  At the moment this is in development so we load the cmos directly.

```ocaml
(* load libraries *)
#use "topfind";;
#require "hardcaml,deriving,mparser,iocaml-kernel.notebook";;
#directory "_build";;
#load "deriving_bits.cmo";;
#load "asm.cmo";;
#load "ttc.cmo";;

(* set up modules *)
module B = HardCaml.Bits.Comb.IntbitsList;;
module A = Asm.Make(B);;
open Ttc;;
open A.Instruction;;
open A.Asm;;

(* install toplevel printers *)
let asm_printer fmt b =
    Iocaml.display "text/html" (A.Printer.html_of_asm_instr b);
    Format.fprintf fmt "<html>"
;;
#install_printer asm_printer;;
let program_printer fmt b =
    Iocaml.display "text/html" (A.Printer.html_of_program b);
    Format.fprintf fmt "<html>"
;;
#install_printer program_printer;;
let assembled_program_printer fmt b =
    Iocaml.display "text/html" (A.Printer.html_of_assembled_program b);
    Format.fprintf fmt "<html>"
;;
#install_printer assembled_program_printer;;
```

# Instruction Format

<img src="http://www.cl.cam.ac.uk/teaching/1314/ECAD+Arch/labs/images/instructions.png"></img>

The general operation of the machine is to read registers `Ra` and `Rb` from a 128 element register file, compute a function of these operands (like `add` or `xor`), rotate the result right by 0, 1, 8 or 16 bits and write the result to `Rw`.

Based on `skip` the following instruction can be optionally skipped depending on the ALU functions result.  This is done by incrementing the program counter (pc) by 2 rather than the default 1.

The special `jump` instruction allows (along with `skip` conditions) arbitrary flow control.

Additionally a number of special instructions deal with external memories and peripherals.

Finally, the `const` instruction loads a register with a 24 bit constant value.

# Assembler

## Low level assembler

This basic assembler API consists of the following 4 functions

```ocaml
(* general instruction format *)
instr
```

```ocaml
(* load a 24 bit constant *)
const
```

```ocaml
(* loads a constant the value of which is calculated 
   by the assembler from a string label.  Used for
   program addresses (ie jumps + branches) *)
const_label
```

```ocaml
(* groups several instructions together.
   Used for simple macro instruction sequences *)
block
```

The `instr` function takes each of the various operation fields in the instruction as optional arugments.  By default these perform the `and` function with no rotate or skip.

```ocaml
block [
    instr ~func:Func.Add 0 1 2;
    instr ~func:Func.Sub ~rot:Rotate.Rcy8 ~skip:Skip.Ltz ~op:Op.Ldd 5 4 3;
    const 3 40;
    const_label 3 "loop"
]
```

## Operator based assembler

A simple layer of operators/functions allows a slightly more natural syntax.  The operator and low level syntaxes can be freely mixed.

```ocaml
block [
    0 <-- (1 +: 2);
    1 <-- (3 &: 19);
    2 <-. 223;
    8 <-- (^:) 127 5 ~rot:Rotate.Rcy8; (* use the operator in prefix form to apply optional arguments *)
    9 <-- inc 3 ~skip:Skip.Eqz;
    0 <-- (&:) 0 0 ~op:Op.Jmp ~skip:Skip.Inr;
    9 <-: "main";
    instr ~func:Func.Add 0 1 2;
]
```

## Text based assembler

A parser for an assembler [grammer](http://www.cl.cam.ac.uk/teaching/1314/ECAD+Arch/labs/files/syntax.txt) is also implemented.

```ocaml
A.Parser.parse_string "
    add r0 r1 r2 rot8
    jmp r0 r1
    inc r0 sez
"
```

# Programs

A program is a list of instructions and labels as follows.

```ocaml
let program = [
    "main", [ 
        0 <-. 20;
        1 <-. 40;
        2 <-- (0 +: 1); 
    ];
    "loop", [ 
        3 <-- (2 &: 0);
        0 <-- jmp 0 0;
        block [
            1 <-- jmpl "loop";
            2 <-: "main";
        ]
    ]
]
```

Programs can then be assembled to binary code.

```ocaml
let bin = assemble program
```

And then disassembled.  Labels are lost during the assembly process but we can see the constants they resolved to within the program code.

```ocaml
disassemble bin
```

# Example program

This program calculates the sum of the integers between zero and a value taken from the input stream.  The result is put on the output stream.

```ocaml
let sum_asm = "
    lc r1 main
main:
    ldin r2 sin                    # Load a value from the input stream into r2...
    jmp r0 r1                      # ...but repeat if the stream was not ready (sin)
    
    lc r3 0                        # Initialise result register
    lc r4 loop
    
loop:                       
    add r3 r3 r2 sez               # Add the current number to the result...
    dec r2                         # ...and decrement if non-zero (sez)
    bnez r0 r2 r4                  # Repeat if the current number is non-zero
    
    stout r3                       # Otherwise, put the result on the output stream

    lc r1 end
end:
    jmp r0 r1                      # Halt
" 

let sum_program_asm = assemble (match A.Parser.parse_string sum_asm with Some(x) -> x | None -> [])
```

The same thing written with the OCaml API.

```ocaml
let sum_ast = [
    "", [
        1 <-: "main";
    ];
    "main", [
        instr ~op:Op.In ~skip:Skip.Inr 2 0 0;
        0 <-- jmp 1 1;
        3 <-. 0;
        4 <-: "loop";
    ];
    "loop", [
        3 <-- (+:) 3 2 ~skip:Skip.Eqz;
        2 <-- dec 2;
        0 <-- bnez 2 4;
        instr ~op:Op.Out 3 3 3;
        1 <-: "end";
    ];
    "end", [
        0 <-- jmp 1 1;
    ];
]
let sum_program_ast = assemble sum_ast
```

# Interpreter

The interpreter can run an assembled program.  It takes 5 parameters;

* start_address - initial value of the program counter (usually 0)
* im - instruction memory (assembled program)
* dm - data memory (if any)
* input - function called for an `In` operation
* output - function called for an `Out` operation

It returns an object with functions to step the machine and query it's state.

```ocaml
(* a little bit of plumbing.  Output from the program goes to this cell 
   (so we can see it while stepping) *)
let context = Iocaml.cell_context()

(* create the interpreter *)
let interp = A.Interpreter.make 
    ~start_address:0
    ~input:(Some(fun _ _ -> 10l))
    ~output:(Some(fun a -> Iocaml.display ~context "text/html" (Printf.sprintf "<b>OUTPUT: %li<b>" a)))
    ~im:sum_program_ast
    ~dm:[||]

(* show state, and step *)
let state() = Iocaml.display "text/html" (A.Printer.html_of_interpreter sum_program_ast interp)
```

Each call to `step` will execute one instruction and display the machine state.  Hitting _ctrl-enter_ below repeatedly will execute the program.

```ocaml
interp#step; state()
```

# Hardware simulation

The hardware model exposes a similar interface to the interpreter.

If the gtwave waveform viewer is installed a trace of the simulation can be shown.

```ocaml
module Hw = Ttc.Make(B)
let hw = Hw.sim ~gtkwave:true ~input:None ~output:None ~im:sum_program_ast ~dm:[||]
let state() = Iocaml.display "text/html" (A.Printer.html_of_interpreter sum_program_ast hw)
```

```ocaml
hw#step; state()
```
